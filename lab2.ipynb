```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ==========================================
# Частина 1: NumPy (CRUD операції)
# ==========================================
print("--- Частина 1: NumPy ---")


vector_data = np.arange(0, 20, 2) 
matrix_data = np.random.rand(3, 3)
scalar_data = np.array(42)

print("Створений вектор:", vector_data)
print("Створена матриця:\n", matrix_data)


print(f"\nЧитання 3-го елементу вектору: {vector_data[2]}")
print(f"Читання елементу матриці (ряд 1, кол 1): {matrix_data[1, 1]}")


vector_data[2] = 999  
print(f"Оновлений вектор: {vector_data}")


vector_data = np.delete(vector_data, 2)
print(f"Вектор після видалення елементу: {vector_data}")


# ==========================================
# Частина 2: Pandas (DataFrame та маніпуляції)
# ==========================================
print("\n--- Частина 2: Pandas ---")


data_dict = {
    'Name': ['Anna', 'Borys', 'Clara', 'Denis', 'Elena'],
    'Age': [21, 25, 23, 30, 22],
    'Score': [88, 92, 79, 95, 85]
}
df_dict = pd.DataFrame(data_dict)


numpy_arr = np.random.randint(0, 100, size=(5, 3))
df_numpy = pd.DataFrame(numpy_arr, columns=['A', 'B', 'C'])


print("\nПерші рядки (head):")
print(df_dict.head(3))  

print("\nСтатистичний опис (describe):")
print(df_dict.describe()) 

print("\nВибір за індексом (iloc) - другий рядок:")
print(df_dict.iloc[1])    

print("\nВибір за міткою (loc) - вік першої людини:")
print(df_dict.loc[0, 'Age']) 


# ==========================================
# Частина 3: Matplotlib (Графік та метрики)
# ==========================================
print("\n--- Частина 3: Matplotlib та Метрики ---")


m = 2.5  
b = 10   
n_points = 50

X = np.linspace(0, 10, n_points)
Y_ideal = m * X + b  


noise = np.random.normal(0, 2, n_points) 
Y_hat = Y_ideal + noise  


mae = np.mean(np.abs(Y_ideal - Y_hat))
mse = np.mean((Y_ideal - Y_hat)**2)

print(f"MAE (Середня абсолютна похибка): {mae:.4f}")
print(f"MSE (Середня квадратична похибка): {mse:.4f}")


plt.figure(figsize=(10, 6))
plt.plot(X, Y_ideal, color='red', label=f'Лінійна функція y={m}x+{b}', linewidth=2)
plt.scatter(X, Y_hat, color='blue', label='Дані з шумом (вимірювання)', alpha=0.6)

plt.title('Графік лінійної функції та згенерованих даних')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()


# ==========================================
# Частина 4: Збереження у CSV
# ==========================================
print("\n--- Частина 4: Експорт результатів ---")


results_df = pd.DataFrame({
    'X': X,
    'Y_ideal': Y_ideal,
    'Y_hat': Y_hat,
    'MAE': mae,      
    'MSE': mse        
})


filename = 'Lab2_results.csv'
results_df.to_csv(filename, index=False)
print(f"Результати успішно записані у файл {filename}")
```
